```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

df = pd.read_csv('../base_dados/beber_copia2.csv', low_memory=False)
```

## MÃ‰TODO DO COTOVELO PARA SABER QUAL MELHOR VALOR PARA CADA ATRIBUTO

#### waistline

```python

data = df[['waistline']].dropna()  # Remover valores ausentes

sse = []

# Intervalo de clusters de 2 a 20
for k in range(2, 21):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    sse.append(kmeans.inertia_)

# Plotar o grÃ¡fico do MÃ©todo do Cotovelo
plt.figure(figsize=(8, 5))
plt.plot(range(2, 21), sse, marker='o')
plt.xlabel('NÃºmero de Clusters (k)')
plt.ylabel('Soma dos Erros QuadrÃ¡ticos (SSE)')
plt.title('MÃ©todo do Cotovelo')
plt.xticks(range(2, 21))
plt.grid()
plt.show()
```

```python

X = df[['waistline']]

# Aplicar o KMeans
kmeans = KMeans(n_clusters=6, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

# Calcular os centrÃ³ides reais dos clusters
centroides_reais = df.groupby('cluster')['waistline'].mean()

# Mapear os valores dos centrÃ³ides formatados para uma nova coluna waistline_grupo
df['waistline_grupo'] = df['cluster'].map(lambda cluster: round(centroides_reais[cluster], 2))

# Remover a coluna cluster, caso nÃ£o seja necessÃ¡ria
df.drop(columns=['cluster'], inplace=True)
df.drop(columns=['waistline'], inplace=True)

```
