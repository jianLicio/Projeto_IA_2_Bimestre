```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

df = pd.read_csv('../base_dados/beber_copia2.csv', low_memory=False)
```

```python
X = df.drop(columns=['drk_yn'])  # Dados preditores
y = df['drk_yn']  # Coluna alvo

# Treinar um modelo de Random Forest para calcular as importncias dos atributos
model = RandomForestClassifier(random_state=42)
model.fit(X, y)

# Obter as importncias dos atributos
importances = model.feature_importances_

# Criar um DataFrame para organizar as importncias
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
})

# Ordenar as caractersticas por importncia
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Exibir o ranking
print(feature_importance_df)



# Criar um grfico de barras para a importncia dos atributos
plt.figure(figsize=(10, 6))
plt.bar(feature_importance_df['Feature'], feature_importance_df['Importance'], color='skyblue')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Feature Importance')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df.drop(columns=['hear_left'], inplace=True)
df.drop(columns=['hear_right'], inplace=True)
```
